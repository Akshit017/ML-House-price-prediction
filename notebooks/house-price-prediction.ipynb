```python
house_data = house_data.dropna(subset=['Distance', 'Propertycount'])
house_data[['Bathroom', 'Car']] = house_data[['Bathroom', 'Car']].fillna(0)
house_data['Bedroom2'].fillna(house_data['Rooms'], inplace=True)
house_data.drop('SaleYear', axis=1, inplace=True)
```

```python
house_data['Landsize'].fillna(house_data['Landsize'].mean(), inplace=True)
house_data['BuildingArea'].fillna(house_data['BuildingArea'].mean(), inplace=True)
```

```python
house_data = house_data.drop(['CouncilArea', 'Regionname'], axis=1)
```

```python
suburb_target_map = house_data.groupby('Suburb')['Price'].mean().to_dict()
house_data['Suburb_encoded'] = house_data['Suburb'].map(suburb_target_map)
house_data.drop('Suburb', axis=1, inplace=True)
```

```python
house_data.isnull().sum()
```

```python
house_data[house_data['Suburb_encoded'].isnull() == True]
```

```python
house_data = house_data.dropna(subset=['Suburb_encoded'])
```

```python
house_data.isnull().sum()
```

```python
house_data = pd.get_dummies(house_data, columns=['Method', 'Type'], drop_first=True)
```

```python
house_data.head()
```

```python
house_data = house_data.astype({col: 'int' for col in house_data.columns if house_data[col].dtype == 'bool'})
```

```python
house_data = house_data.dropna(subset=['Price'])
```

```python
house_data.head()
```

```python
house_data.isnull().sum()
```

```python
house_data = house_data.astype({
    'Rooms': 'int',
    'Bedroom2': 'int',
    'Bathroom': 'int',
    'Car': 'int',
    'Propertycount': 'int',
    'Suburb_encoded': 'int'
})
```

```python
house_data.head(20)
```

```python
house_data = house_data.round({
    'Landsize': 2,
    'BuildingArea': 2
})
```

```python
x = house_data.drop('Price', axis=1)
y = house_data['Price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
```

```python
model.fit(x, y)
```

```python
a = pd.Series(model.feature_importances_, index=x.columns)
a.nlargest(12).plot(kind='barh')
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.svm import SVR
```

```python
allModel = [DecisionTreeRegressor, KNeighborsRegressor, RandomForestRegressor, AdaBoostRegressor, SVR]

for i in allModel:
    model = i()
    model = model.fit(x_train, y_train)
    print(i, 'accuracy', model.score(x_test, y_test))
```

```python
model.score(x_train, y_train)
```
